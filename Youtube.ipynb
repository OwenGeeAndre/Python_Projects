# DATA CLEANING AND WRANGLING

```python
#Removing unwanted columns

netf.drop(['Abbreviation',
       'Gross tertiary education enrollment (%)','rank', 'Unemployment rate','Urban_population','Population','subscribers_for_last_30_days','Latitude','Longitude'],axis=1,inplace=True
       )

netf.columns
```

```python
#Renaming Column Headers

netf.rename(columns = {'Youtuber':'youtuber','Title':'title','video views':'views','Country':'country', 'video_views_for_the_last_30_days':'last_monthly_views'},inplace=True)
```

```python
netf.head(2)
```

```python
netf.isna().sum()
```

```python


cols=['country_rank','video_views_rank','channel_type_rank']
netf[cols] = netf[cols].fillna(0)

netf.fillna(value={'category':'unknown','country':'unspecified',\
                   'channel_type':'unknown'},inplace=True)
```

```python
#Removing Blanks

netf.dropna(subset=['last_monthly_views','created_year','created_month','created_date'],axis=0,inplace=True)

sns.heatmap(netf.isnull())

plt.title('Heatmap for Blanks')

plt.show()
```

```python
#Converting data types of different columns


netf.views = netf.views.astype(np.int64)
netf.last_monthly_views = netf.last_monthly_views.astype(np.int64)


netf['video_views_rank'] = netf['video_views_rank'].astype(str)
netf['country_rank'] = netf['country_rank'].astype(str)
netf['channel_type_rank'] = netf['channel_type_rank'].astype(str)


netf['created_year'] = netf['created_year'].astype(str)
netf['created_date'] = netf['created_date'].astype(str)

```

```python
numnetf = netf.copy()

for cols in numnetf:
    if(numnetf[cols].dtype == 'object'):
        numnetf[cols] = numnetf[cols].astype('category')
        numnetf[cols] = numnetf[cols].cat.codes

numnetf.head()
```

```python
plt.figure(figsize=(15,15))

corr_netf = numnetf.corr(method = 'pearson')

sns.heatmap(corr_netf,annot=True)

plt.title('Correlation for All Features')
plt.xlabel('Youtube Stats')
plt.ylabel('Youtube Stats')

plt.show()
```

```python
corr_matx = corr_netf.corr()
corr_pairs = corr_matx.unstack()
corr_pairs
```

```python
best_netf_corr = corr_pairs[((corr_pairs)>0.7) & ((corr_pairs)<0.99)|((corr_pairs)<-0.7) & ((corr_pairs)>-0.99)]

best_netf_corr
```
